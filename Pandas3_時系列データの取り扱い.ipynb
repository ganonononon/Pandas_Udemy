日べつの売り上げを求める

pandasでDataFrameやSeriesの欠損値NaNを任意の値に置換（穴埋め、代入）するにはfillna()メソッドを使う。

```python
df = df.fillna({'quantity': 1})
df
```

```python
df['sales'] = df['revenue'] * df['quantity']
df
```

```python
daily_sales = df.groupby('ymd')['sales'].sum().reset_index()
daily_sales
```

plt.xticks(rotation=90)
x軸を90度回転

```python
plt.plot(daily_sales['ymd'],daily_sales['sales'])
plt.xticks(rotation=90)
```
