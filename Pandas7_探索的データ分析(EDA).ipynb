データを多面的に分析し、数値に基づいて、ビジネス上有用なインサイトを引き出す事

```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/UdemyPandas/dataset/citibike_trips_2017_sampled.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.info()
```

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
```

```python
df.info()
```

```python
df['num_trips'] = 1
```

```python
df['tripduration_min'] = df['tripduration'] / 60.0
```

```python
df['month'] = df['starttime'].dt.month
df['day'] = df['starttime'].dt.day
df['date'] = df['starttime'].dt.date
df['hour'] = df['starttime'].dt.hour
df['day_of_week'] = df['starttime'].dt.dayofweek
```

```python
df
```

月別の利用者数をカウント

```python
num_trips_df = df.groupby('month')['num_trips'].count().reset_index()
num_trips_df
```

```python
px.line(num_trips_df, x='month', y='num_trips', markers=True,
        width=800,height=600)
```

station別

```python
station_df = df.groupby('start_station_id')['num_trips'].count().reset_index()
station_df
```

```python
top10_stations = station_df.sort_values('num_trips', ascending=False).head(10)['start_station_id']
top10_stations
```

```python
num_trips_df = df.groupby(['start_station_id','day_of_week'])['num_trips'].count().reset_index()
query = 'start_station_id in @top10_stations'
px.line(num_trips_df.query(query), x='day_of_week',y='num_trips',
        color='start_station_id', width=800, height=600)
```

```python
px.histogram(df, x='hour', facet_col='dayofweek')
```

```python
df.columns
```

```python
px.histogram(df, x='hour', facet_col='dayofweek', facet_row='usertype')
```

クロス集計

```python
demogra_pivot = pd.pivot_table(df, index='age', columns ='gender',
                               values='num_trips', aggfunc='sum').fillna(0)
```

```python
demogra_pivot
```

ヒートマップ

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 8))
fig = sns.heatmap(demogra_pivot, annot=True)
```

デモグラ別の利用時間

```python
sns.boxplot(df.query('gender in ["male", "female"]'),
            x='age', y='tripduration_min', hue='gender')
```

```python
fig = plt.figure(figsize=(12, 8))
query = '(gender in ["male", "female"]) & (tripduration_min <= 100) & (age <= 100)'
fig = sns.boxplot(data=df.query(query),
            x='age', y='tripduration_min', hue='gender')
```
